```python
import matplotlib.pyplot as plt
import numpy as np

n = np.array([10, 100, 1000, 10000, 100000, 200000, 300000])
insertion = np.array([9e-06, 6.5e-05, 0.00397, 0.160486, 13.4974, 53.7029, 123.487])
merge     = np.array([6e-05, 0.000111, 0.001418, 0.005429, 0.061544, 0.128986, 0.199978])
quick     = np.array([1e-06, 1.2e-05, 0.000219, 0.001055, 0.012035, 0.036277, 0.040789])
```

```python
plt.plot(n, insertion, label="Insertion Sort")
plt.plot(n, quick, label="Quick Sort")
plt.plot(n, merge, label="Merge Sort")
plt.xlabel('vector size')
plt.ylabel('time in seconds')
plt.grid(True)
plt.legend()
plt.savefig("assignment1_plot1.png")
```

```python
plt.plot(n, quick, label="Quick Sort")
plt.plot(n, merge, label="Merge Sort")
plt.xlabel('vector size')
plt.ylabel('time in seconds')
plt.grid(True)
plt.legend()
plt.savefig("assignment1_plot2.png")
```
